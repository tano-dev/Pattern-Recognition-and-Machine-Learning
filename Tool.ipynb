## üéØ 3. Ch·ªçn & L·ªçc D·ªØ Li·ªáu (Selection & Filtering)
* **Ch·ªçn C·ªôt:**
    * `df['col']` ho·∫∑c `df.col`: Ch·ªçn m·ªôt c·ªôt (tr·∫£ v·ªÅ Series).
    * `df[['col1', 'col2']]`: Ch·ªçn nhi·ªÅu c·ªôt (tr·∫£ v·ªÅ DataFrame).
* **Ch·ªçn D√≤ng (theo nh√£n/v·ªã tr√≠):**
    * `df.loc[label]`: Ch·ªçn m·ªôt d√≤ng theo nh√£n index (v√≠ d·ª•: `df.loc[0]`).
    * `df.loc['label1':'label5']`: Ch·ªçn m·ªôt kho·∫£ng d√≤ng theo nh√£n (slice, bao g·ªìm c·∫£ ƒëi·ªÉm cu·ªëi).
    * `df.iloc[v·ªã_tr√≠]`: Ch·ªçn m·ªôt d√≤ng theo v·ªã tr√≠ s·ªë nguy√™n (v√≠ d·ª•: `df.iloc[0]`).
    * `df.iloc[0:5]`: Ch·ªçn 5 d√≤ng ƒë·∫ßu ti√™n (slice, kh√¥ng bao g·ªìm ƒëi·ªÉm cu·ªëi, gi·ªëng Python list).
* **Ch·ªçn D√≤ng & C·ªôt:**
    * `df.loc[row_labels, col_labels]`: Ch·ªçn theo nh√£n.
        * *V√≠ d·ª•:* `df.loc[0:5, ['name', 'age']]`

## üîß 4. L√†m S·∫°ch & Bi·∫øn ƒê·ªïi D·ªØ Li·ªáu (Cleaning & Transformation)

    * `df.dropna(axis=1)`: X√≥a t·∫•t c·∫£ c√°c c·ªôt c√≥ ch·ª©a NaN.

    * `df.rename(columns={'old_name': 'new_name'})`: ƒê·ªïi t√™n c·ªôt.
    * `df.replace(old_value, new_value)`: Thay th·∫ø gi√° tr·ªã.
    * `df['col'].astype('new_type')`: Thay ƒë·ªïi ki·ªÉu d·ªØ li·ªáu c·ªßa c·ªôt (v√≠ d·ª•: `'int'`, `'float'`, `'category'`).
* **X√≥a D√≤ng/C·ªôt:**

    * `df.drop(labels=[index_label])`: X√≥a d√≤ng theo index.

    * `df.drop_duplicates(subset=['col_name'], keep='first')`: X√≥a c√°c d√≤ng tr√πng l·∫∑p (ch·ªâ gi·ªØ l·∫°i d√≤ng ƒë·∫ßu ti√™n).

## üîó 7. H·ª£p Nh·∫•t & K·∫øt N·ªëi (Merge & Concat)
* `pd.concat([df1, df2])`: N·ªëi c√°c DataFrame theo chi·ªÅu d·ªçc (stack rows, `axis=0`).
* `pd.concat([df1, df2], axis=1)`: N·ªëi c√°c DataFrame theo chi·ªÅu ngang (gh√©p c·ªôt, `axis=1`).
* `pd.merge(df_left, df_right, on='key_col')`: K·∫øt n·ªëi (join) hai DataFrame d·ª±a tr√™n m·ªôt c·ªôt chung (key).
    * *Tham s·ªë `how` (c√°ch join):*
        * `how='inner'` (m·∫∑c ƒë·ªãnh): Ch·ªâ gi·ªØ l·∫°i c√°c h√†ng c√≥ key chung ·ªü c·∫£ hai b·∫£ng.

## 1. üìñ Gi·ªõi Thi·ªáu
* **X√°c Su·∫•t**: ƒê·∫°i l∆∞·ª£ng d√πng ƒë·ªÉ ƒëo kh·∫£ nƒÉng x·∫£y ra c·ªßa m·ªôt s·ª± ki·ªán (event), c√≥ gi√° tr·ªã t·ª´ 0 ƒë·∫øn 1.
* **Ph√¢n Ph·ªëi X√°c Su·∫•t (PPXS)**: Th·ªÉ hi·ªán x√°c su·∫•t xu·∫•t hi·ªán c·ªßa c√°c gi√° tr·ªã c·ªßa m·ªôt bi·∫øn ng·∫´u nhi√™n. Gi√∫p hi·ªÉu quy lu·∫≠t c·ªßa d·ªØ li·ªáu, d·ª± ƒëo√°n gi√° tr·ªã, so s√°nh d·ªØ li·ªáu v·ªõi c√°c PPXS ƒë√£ bi·∫øt.
* **Bi·∫øn Ng·∫´u Nhi√™n (BNN):** C√°c ƒë·∫∑c tr∆∞ng d·ªØ li·ªáu ƒë∆∞·ª£c gi·∫£ ƒë·ªãnh l√† bi·∫øn ng·∫´u nhi√™n.
    * **R·ªùi r·∫°c (Discrete):** Nh·∫≠n c√°c gi√° tr·ªã ƒë·∫øm ƒë∆∞·ª£c (vd: s·ªë l·∫ßn tung x√∫c x·∫Øc, s·ªë c√¢u ƒë√∫ng, s·ªë kh√°ch h√†ng). M√¥ t·∫£ b·∫±ng **H√†m kh·ªëi x√°c su·∫•t (PMF)**.
    * **Li√™n t·ª•c (Continuous):** Nh·∫≠n c√°c gi√° tr·ªã trong m·ªôt kho·∫£ng (vd: chi·ªÅu cao, nhi·ªát ƒë·ªô, th·ªùi gian). M√¥ t·∫£ b·∫±ng **H√†m m·∫≠t ƒë·ªô x√°c su·∫•t (PDF)**.

## 3. ‚öôÔ∏è C√°c H√†m & Tham S·ªë Chung

### A. C√°c H√†m T√≠nh X√°c Su·∫•t & L·∫•y M·∫´u (Function Details)
* `<dist>.rvs(...)`: T·∫°o m·∫´u ng·∫´u nhi√™n (Random Variates).
    * `size`: S·ªë l∆∞·ª£ng m·∫´u c·∫ßn t·∫°o.
    * C√°c tham s·ªë ƒë·∫∑c tr∆∞ng c·ªßa ph√¢n ph·ªëi (vd: `n`, `p` cho `binom`).
* `<dist>.pmf(k, ...)`: **H√†m kh·ªëi x√°c su·∫•t** (Probability Mass Function).
    * D√πng cho BNN **R·ªùi r·∫°c**.
    * T√≠nh x√°c su·∫•t $P(X=k)$. (H√†m ƒë·ªô l·ªõn f(x) cho ph√¢n ph·ªëi r·ªùi r·∫°c).
* `<dist>.pdf(x, ...)`: **H√†m m·∫≠t ƒë·ªô x√°c su·∫•t** (Probability Density Function).
    * D√πng cho BNN **Li√™n t·ª•c**.
    * $f(x)$ (H√†m m·∫≠t ƒë·ªô f(x) cho ph√¢n ph·ªëi li√™n t·ª•c. L∆∞u √Ω: $f(x) \ge 0$ v√† t√≠ch ph√¢n to√†n mi·ªÅn b·∫±ng 1, $f(x)$ kh√¥ng ph·∫£i l√† x√°c su·∫•t t·∫°i ƒëi·ªÉm x).
* `<dist>.cdf(x, ...)`: **H√†m ph√¢n ph·ªëi t√≠ch l≈©y** (Cumulative Distribution Function).
    * T√≠nh x√°c su·∫•t $P(X \le x)$ (H√†m F(x)).
* `<dist>.ppf(q, ...)`: **H√†m ƒëi·ªÉm ph·∫ßn trƒÉm** (Percent Point Function).
    * H√†m ng∆∞·ª£c c·ªßa CDF (Inverse of CDF), c√≤n g·ªçi l√† Quantile Function.
    * T√¨m $x$ sao cho $P(X \le x) = q$ (v·ªõi $q$ l√† x√°c su·∫•t). (T√¨m ƒëi·ªÉm ph√¢n v·ªã).
* `<dist>.sf(x, ...)`: **H√†m s·ªëng s√≥t** (Survival Function).
    * T√≠nh x√°c su·∫•t $P(X > x)$ (b·∫±ng `1 - cdf(x)`). (H√†m Survival).
* `<dist>.isf(q, ...)`: **H√†m ng∆∞·ª£c c·ªßa h√†m s·ªëng s√≥t** (Inverse Survival Function).
    * T√¨m $x$ sao cho $P(X > x) = q$. (H√†m ng∆∞·ª£c c·ªßa h√†m Survival).

### B. C√°c Tham S·ªë Chung (loc, scale)
H·∫ßu h·∫øt c√°c ph√¢n ph·ªëi li√™n t·ª•c ƒë·ªÅu c√≥ 2 tham s·ªë chu·∫©n h√≥a:
* `loc`: (Location) D·ªãch chuy·ªÉn v·ªã tr√≠ ph√¢n ph·ªëi.
    * V·ªõi `norm`, `loc` ch√≠nh l√† trung b√¨nh $\mu$.
    * V·ªõi `uniform`, `loc` l√† ƒëi·ªÉm b·∫Øt ƒë·∫ßu $a$.
* `scale`: (Scale) Co d√£n ph√¢n ph·ªëi (quy ƒë·ªãnh ƒë·ªô r·ªông).
    * V·ªõi `norm`, `scale` ch√≠nh l√† ƒë·ªô l·ªách chu·∫©n $\sigma$.
    * V·ªõi `uniform`, `scale` l√† ƒë·ªô r·ªông $b - a$.
    * V·ªõi `expon`, `scale` l√† $1/\lambda$ (ngh·ªãch ƒë·∫£o c·ªßa rate $\lambda$, t·ª©c l√† gi√° tr·ªã trung b√¨nh).

### C. C√°c H√†m Th·ªëng K√™
* `<dist>.mean(...)`: Tr·∫£ v·ªÅ k·ª≥ v·ªçng (trung b√¨nh) c·ªßa ph√¢n ph·ªëi.
* `<dist>.median(...)`: Tr·∫£ v·ªÅ trung v·ªã.
* `<dist>.var(...)`: Tr·∫£ v·ªÅ ph∆∞∆°ng sai ($\sigma^2$).
* `<dist>.std(...)`: Tr·∫£ v·ªÅ ƒë·ªô l·ªách chu·∫©n ($\sigma$).

## 4. Ph√¢n Ph·ªëi Nh·ªã Th·ª©c (`binom`) - R·ªùi R·∫°c
* **K√Ω hi·ªáu**: $B(n, p)$
    * `n`: S·ªë l·∫ßn th·ª±c hi·ªán ph√©p th·ª≠ Bernoulli ƒë·ªôc l·∫≠p.
    * `p`: X√°c su·∫•t th√†nh c√¥ng trong m·ªôt l·∫ßn th·ª≠.
* **Bi·∫øn ng·∫´u nhi√™n X**: S·ªë l·∫ßn th√†nh c√¥ng trong `n` ph√©p th·ª≠.
* **V√≠ d·ª•**: M·ªôt b√†i thi tr·∫Øc nghi·ªám g·ªìm 10 c√¢u, m·ªói c√¢u c√≥ 4 l·ª±a ch·ªçn (p=0.25). B·∫°n ƒë√°nh l·ª•i t·∫•t c·∫£. X l√† s·ªë c√¢u tr·∫£ l·ªùi ƒë√∫ng. $X \sim B(10, 0.25)$.

````python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

n, p = 10, 0.25

# 1. Ph√°t sinh m·∫´u ng·∫´u nhi√™n (rvs)
# Ph√°t sinh 1 gi√° tr·ªã (1 b·∫°n l√†m b√†i)
mau_1 = binom.rvs(n, p)
print(f"M·ªôt k·∫øt qu·∫£ ng·∫´u nhi√™n: {mau_1}")

# Ph√°t sinh 1000 gi√° tr·ªã (1000 b·∫°n l√†m b√†i)
mau_1000 = binom.rvs(n, p, size=1000)

# V·∫Ω histogram c·ªßa m·∫´u
plt.figure(figsize=(10, 5))
dorongcot = 0.4
offset = dorongcot / 2
plt.hist(mau_1000, bins=np.arange(n + 2) - offset, width=dorongcot, density=True, label='Histogram t·ª´ m·∫´u RVS')
plt.xticks(range(n + 1))
plt.xlabel('S·ªë c√¢u ƒë√∫ng (k)')
plt.ylabel('T·∫ßn su·∫•t t∆∞∆°ng ƒë·ªëi / M·∫≠t ƒë·ªô')
plt.title(f'Histogram Ph√¢n ph·ªëi Nh·ªã th·ª©c (n={n}, p={p}) t·ª´ 1000 m·∫´u')

# 2. T√≠nh v√† v·∫Ω h√†m PMF (x√°c su·∫•t l√Ω thuy·∫øt P(X=k))
k_values = np.arange(0, n + 1)
pmf_values = binom.pmf(k_values, n, p)
plt.stem(k_values, pmf_values, linefmt='r-', markerfmt='ro', basefmt=' ', label='PMF L√Ω thuy·∫øt')
plt.legend()
plt.show()

# 3. T√≠nh v√† v·∫Ω h√†m CDF (x√°c su·∫•t t√≠ch l≈©y P(X <= k))
plt.figure(figsize=(10, 5))
cdf_values = binom.cdf(k_values, n, p)
plt.plot(k_values, cdf_values, marker='o', drawstyle='steps-post', label='CDF') # steps-post cho ƒë·ªì th·ªã b·∫≠c thang
plt.title(f'H√†m Ph√¢n ph·ªëi T√≠ch l≈©y (CDF) Nh·ªã th·ª©c (n={n}, p={p})')
plt.xlabel('S·ªë c√¢u ƒë√∫ng (k)')
plt.ylabel('P(X <= k)')
plt.xticks(k_values)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()

# 4. T√≠nh to√°n x√°c su·∫•t c·ª• th·ªÉ
# P(X = 3) - X√°c su·∫•t ƒë√∫ng ch√≠nh x√°c 3 c√¢u
p_eq_3 = binom.pmf(3, n, p)
print(f"P(X = 3) = {p_eq_3:.4f}")
# P(X <= 2) - X√°c su·∫•t ƒë√∫ng t·ªëi ƒëa 2 c√¢u
p_le_2 = binom.cdf(2, n, p)
print(f"P(X <= 2) = {p_le_2:.4f}")
# P(X > 5) - X√°c su·∫•t ƒë√∫ng nhi·ªÅu h∆°n 5 c√¢u
p_gt_5 = binom.sf(5, n, p) # <=> 1 - binom.cdf(5, n, p)
print(f"P(X > 5) = {p_gt_5:.4f}")
# T√¨m k nh·ªè nh·∫•t sao cho P(X <= k) >= 0.9 (Ph√¢n v·ªã th·ª© 90)
k_q90 = binom.ppf(0.9, n, p)
print(f"Ph√¢n v·ªã th·ª© 90 (k sao cho P(X<=k)>=0.9): {k_q90}")
````

## 5\. Ph√¢n Ph·ªëi Poisson (`poisson`) - R·ªùi R·∫°c

  * **K√Ω hi·ªáu**: $P(\lambda)$
      * `lambda` ($\lambda$): Trung b√¨nh s·ªë l·∫ßn x·∫£y ra s·ª± ki·ªán trong m·ªôt kho·∫£ng (th·ªùi gian, kh√¥ng gian). Trong `scipy` tham s·ªë n√†y t√™n l√† `mu`.
  * **Bi·∫øn ng·∫´u nhi√™n X**: S·ªë l·∫ßn s·ª± ki·ªán x·∫£y ra.
  * **V√≠ d·ª•**: S·ªë kh√°ch h√†ng ƒë·∫øn si√™u th·ªã trong 1 gi·ªù (trung b√¨nh $\lambda=20$).

<!-- end list -->

```python
from scipy.stats import poisson

# Trung b√¨nh 20 kh√°ch/gi·ªù
mu_lambda = 20

# P(X = 20) - X√°c su·∫•t c√≥ ƒë√∫ng 20 kh√°ch
p_eq_20 = poisson.pmf(20, mu=mu_lambda)
print(f"P(X = 20) = {p_eq_20:.4f}")
# P(X <= 15) - X√°c su·∫•t c√≥ t·ªëi ƒëa 15 kh√°ch
p_le_15 = poisson.cdf(15, mu=mu_lambda)
print(f"P(X <= 15) = {p_le_15:.4f}")
# P(X > 30) - X√°c su·∫•t c√≥ h∆°n 30 kh√°ch
p_gt_30 = poisson.sf(30, mu=mu_lambda) # <=> 1 - poisson.cdf(30, mu=mu_lambda)
print(f"P(X > 30) = {p_gt_30:.4f}")

# V·∫Ω PMF v√† CDF t∆∞∆°ng t·ª± nh∆∞ Binomial
k_values_poi = np.arange(0, 41) # Ch·ªçn kho·∫£ng gi√° tr·ªã h·ª£p l√Ω quanh lambda
pmf_poi = poisson.pmf(k_values_poi, mu=mu_lambda)
cdf_poi = poisson.cdf(k_values_poi, mu=mu_lambda)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].stem(k_values_poi, pmf_poi, linefmt='b-', markerfmt='bo', basefmt=' ')
axes[0].set(xlabel='S·ªë kh√°ch (k)', title=f'Poisson PMF ($\lambda$={mu_lambda})')
axes[1].plot(k_values_poi, cdf_poi, marker='.', drawstyle='steps-post')
axes[1].set(xlabel='S·ªë kh√°ch (k)', title=f'Poisson CDF ($\lambda$={mu_lambda})')
plt.tight_layout()
plt.show()

## 6\. Ph√¢n Ph·ªëi Chu·∫©n (`norm`) - Li√™n T·ª•c

  * **K√Ω hi·ªáu**: $N(\mu, \sigma^2)$
      * `loc` ($\mu$): K·ª≥ v·ªçng (trung b√¨nh).
      * `scale` ($\sigma$): ƒê·ªô l·ªách chu·∫©n. (L∆∞u √Ω: tham s·ªë l√† $\sigma$, kh√¥ng ph·∫£i $\sigma^2$).
  * **Bi·∫øn ng·∫´u nhi√™n X**: Bi·∫øn li√™n t·ª•c, ƒë·ªì th·ªã h√¨nh chu√¥ng ƒë·ªëi x·ª©ng qua $\mu$.
  * **V√≠ d·ª•**: Chi·ªÅu cao nam thanh ni√™n tr∆∞·ªüng th√†nh ($\mu=170$ cm, $\sigma=3$ cm). $X \sim N(170, 3^2)$.

<!-- end list -->

````python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 170, 3

# 1. Ph√°t sinh m·∫´u ng·∫´u nhi√™n (rvs) v√† v·∫Ω histogram
plt.figure(figsize=(10, 5))
mau_10k = norm.rvs(loc=mu, scale=sigma, size=10000)
plt.hist(mau_10k, bins=100, density=True, alpha=0.6, color='g', label='Histogram t·ª´ m·∫´u RVS')

# 2. T√≠nh v√† v·∫Ω h√†m PDF (h√†m m·∫≠t ƒë·ªô x√°c su·∫•t) l√Ω thuy·∫øt
x_values = np.linspace(mu - 4*sigma, mu + 4*sigma, 200) # Kho·∫£ng gi√° tr·ªã quanh mu (~99.99%)
pdf_values = norm.pdf(x_values, loc=mu, scale=sigma)
plt.plot(x_values, pdf_values, color='red', lw=2, label=f'PDF L√Ω thuy·∫øt $N({mu}, {sigma}^2)$')
plt.title(f'Ph√¢n ph·ªëi Chu·∫©n (Normal) - PDF v√† Histogram M·∫´u')
plt.xlabel('Chi·ªÅu cao (cm)')
plt.ylabel('M·∫≠t ƒë·ªô')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 3. T√≠nh to√°n v·ªõi Ph√¢n ph·ªëi Chu·∫©n T·∫Øc (Z ~ N(0, 1))
print("\n--- T√≠nh to√°n v·ªõi Ph√¢n ph·ªëi Chu·∫©n T·∫Øc Z ~ N(0, 1) ---")
# P(Z <= -1)
p_le_neg1 = norm.cdf(-1) # M·∫∑c ƒë·ªãnh loc=0, scale=1
print(f"P(Z <= -1) = {p_le_neg1:.5f}")
# P(Z > 1) = 1 - P(Z <= 1)
p_gt_1 = 1 - norm.cdf(1)
# Ho·∫∑c d√πng h√†m sf: P(Z > 1)
p_gt_1_sf = norm.sf(1)
print(f"P(Z > 1) = {p_gt_1:.5f} (d√πng 1-cdf) = {p_gt_1_sf:.5f} (d√πng sf)")
# P(-1 <= Z <= 1)
p_between = norm.cdf(1) - norm.cdf(-1)
print(f"P(-1 <= Z <= 1) = {p_between:.5f}")

# 4. T√¨m ƒëi·ªÉm ph√¢n v·ªã (Quantile) v·ªõi Ph√¢n ph·ªëi Chu·∫©n T·∫Øc
print("\n--- T√¨m ƒëi·ªÉm ph√¢n v·ªã (Quantile) v·ªõi Z ~ N(0, 1) ---")
# T√¨m z sao cho P(Z <= z) = 0.02275
z_val_ppf = norm.ppf(0.02275)
print(f"z sao cho P(Z <= z) = 0.02275 l√†: {z_val_ppf:.3f}") # G·∫ßn b·∫±ng -2
# T√¨m z sao cho P(Z > z) = 0.02275
z_val_isf = norm.isf(0.02275)
print(f"z sao cho P(Z > z) = 0.02275 l√†: {z_val_isf:.3f}") # G·∫ßn b·∫±ng 2

# 5. T√≠nh to√°n v·ªõi ph√¢n ph·ªëi chu·∫©n b·∫•t k·ª≥ (v√≠ d·ª• X ~ N(16, 4^2))
print("\n--- T√≠nh to√°n v·ªõi Ph√¢n ph·ªëi Chu·∫©n X ~ N(16, 4^2) ---")
mu_x, sigma_x = 16, 4
# P(X <= 8)
p_le_8 = norm.cdf(8, loc=mu_x, scale=sigma_x)
print(f"P(X <= 8) = {p_le_8:.5f}")
# T√¨m x sao cho P(X <= x) = 0.02275
x_val_ppf = norm.ppf(0.02275, loc=mu_x, scale=sigma_x)
print(f"x sao cho P(X <= x) = 0.02275 l√†: {x_val_ppf:.3f}") # = 16 + (-2)*4 = 8

# V·∫Ω minh h·ªça P(-1 <= Z <= 1)
plt.figure(figsize=(8, 4))
x_z = np.linspace(-4, 4, 200)
pdf_z = norm.pdf(x_z)
plt.plot(x_z, pdf_z, label='PDF c·ªßa Z ~ N(0,1)')
x_fill = np.linspace(-1, 1, 100)
plt.fill_between(x_fill, norm.pdf(x_fill), color='blue', alpha=0.35, label='P(-1 <= Z <= 1)')
plt.title('Minh h·ªça di·ªán t√≠ch d∆∞·ªõi ƒë∆∞·ªùng cong PDF chu·∫©n t·∫Øc')
plt.text(-0.4, 0.1, f'{p_between:.3f}', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

## 7\. Ph√¢n Ph·ªëi ƒê·ªÅu (`uniform`) - Li√™n T·ª•c

  * **K√Ω hi·ªáu**: $U(a, b)$
      * `a`: Gi√° tr·ªã nh·ªè nh·∫•t.
      * `b`: Gi√° tr·ªã l·ªõn nh·∫•t.
  * **√ù nghƒ©a**: M·ªçi gi√° tr·ªã trong kho·∫£ng $[a, b]$ ƒë·ªÅu c√≥ c√πng kh·∫£ nƒÉng x·∫£y ra (h√†m m·∫≠t ƒë·ªô l√† h·∫±ng s·ªë trong kho·∫£ng n√†y).
  * **Tham s·ªë `scipy`**: `loc = a`, `scale = b - a`.

<!-- end list -->

````python
from scipy.stats import uniform

a, b = 10, 30
# loc = 10, scale = 30 - 10 = 20
dist_u = uniform(loc=a, scale=(b-a))

# P(X <= 15)
p_u_le_15 = dist_u.cdf(15)
print(f"P(X <= 15) = {p_u_le_15:.2f}") # (15-10) / (30-10) = 0.25
# P(20 <= X <= 25)
p_u_between = dist_u.cdf(25) - dist_u.cdf(20)
print(f"P(20 <= X <= 25) = {p_u_between:.2f}") # (25-20) / (30-10) = 0.25

# L·∫•y m·∫´u v√† v·∫Ω histogram
mau_deu = dist_u.rvs(10000)
plt.hist(mau_deu, bins=20, density=True, alpha=0.6, label='Histogram M·∫´u')
# V·∫Ω PDF l√Ω thuy·∫øt
x_u = np.linspace(a - 5, b + 5, 100) # V·∫Ω r·ªông h∆°n kho·∫£ng [a,b]
pdf_u = dist_u.pdf(x_u)
plt.plot(x_u, pdf_u, 'r-', lw=2, label='PDF L√Ω thuy·∫øt')
plt.title(f'Ph√¢n ph·ªëi ƒê·ªÅu U({a},{b})')
plt.legend()
plt.show()
````

## 8\. Ph√¢n Ph·ªëi M≈© (`expon`) - Li√™n T·ª•c

  * **K√Ω hi·ªáu**: $Exp(\lambda)$
      * $\lambda$: T·ªëc ƒë·ªô (rate) trung b√¨nh c·ªßa s·ª± ki·ªán (v√≠ d·ª•: s·ªë s·ª± ki·ªán / ƒë∆°n v·ªã th·ªùi gian).
  * **√ù nghƒ©a**: M√¥ t·∫£ th·ªùi gian ch·ªù cho ƒë·∫øn khi s·ª± ki·ªán ti·∫øp theo x·∫£y ra (trong qu√° tr√¨nh Poisson).
  * **Tham s·ªë `scipy`**: `scale = 1 / \lambda`.
      * **Ch√∫ √Ω**: `scale` ch√≠nh l√† **th·ªùi gian ch·ªù trung b√¨nh** gi·ªØa c√°c s·ª± ki·ªán. N·∫øu trung b√¨nh th·ªùi gian ch·ªù l√† 5 ph√∫t ($1/\lambda = 5$), th√¨ `scale=5`.

<!-- end list -->

````python
from scipy.stats import expon

# V√≠ d·ª•: Trung b√¨nh th·ªùi gian ch·ªù xe bus l√† 10 ph√∫t.
# 1/lambda = 10 (th·ªùi gian ch·ªù trung b√¨nh)
# Ta d√πng scale = 10
avg_wait_time = 10
dist_e = expon(scale=avg_wait_time) # loc=0 l√† m·∫∑c ƒë·ªãnh (th·ªùi gian kh√¥ng √¢m)

# P(X <= 5) - X√°c su·∫•t ch·ªù t·ªëi ƒëa 5 ph√∫t
p_e_le_5 = dist_e.cdf(5)
print(f"P(X <= 5) = {p_e_le_5:.4f}")
# P(X > 10) - X√°c su·∫•t ch·ªù h∆°n 10 ph√∫t (h∆°n th·ªùi gian trung b√¨nh)
p_e_gt_10 = dist_e.sf(10)
print(f"P(X > 10) = {p_e_gt_10:.4f}") # Kho·∫£ng 0.368 (1/e)

# V·∫Ω PDF v√† CDF
x_e = np.linspace(0, avg_wait_time * 4, 200) # V·∫Ω ƒë·∫øn kho·∫£ng 4 l·∫ßn trung b√¨nh
pdf_e = dist_e.pdf(x_e)
cdf_e = dist_e.cdf(x_e)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(x_e, pdf_e, label=f'PDF ($1/\lambda$={avg_wait_time})')
axes[0].set(xlabel='Th·ªùi gian ch·ªù (ph√∫t)', title='Exponential PDF')
axes[0].legend()
axes[0].grid(True, linestyle='--', alpha=0.7)
axes[1].plot(x_e, cdf_e, label=f'CDF ($1/\lambda$={avg_wait_time})')
axes[1].set(xlabel='Th·ªùi gian ch·ªù (ph√∫t)', title='Exponential CDF')
axes[1].legend()
axes[1].grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

## 9\. Ph√¢n Ph·ªëi Student (`t`) - Li√™n T·ª•c

  * **K√Ω hi·ªáu**: $t(df)$
      * `df`: B·∫≠c t·ª± do (Degrees of Freedom).
  * **√ù nghƒ©a**: T∆∞∆°ng t·ª± ph√¢n ph·ªëi Chu·∫©n nh∆∞ng c√≥ "ƒëu√¥i n·∫∑ng" (heavy tails), nghƒ©a l√† c√°c gi√° tr·ªã c·ª±c ƒëoan (xa trung b√¨nh) d·ªÖ x·∫£y ra h∆°n, ƒë·∫∑c bi·ªát khi `df` nh·ªè.
  * **S·ª≠ d·ª•ng**: Khi ∆∞·ªõc l∆∞·ª£ng trung b√¨nh t·ªïng th·ªÉ t·ª´ m·∫´u nh·ªè (th∆∞·ªùng $n < 30$) v√† ch∆∞a bi·∫øt ph∆∞∆°ng sai $\sigma^2$ t·ªïng th·ªÉ. Khi `df` l·ªõn (v√≠ d·ª• `df > 30`), ph√¢n ph·ªëi `t` x·∫•p x·ªâ ph√¢n ph·ªëi `norm`.

<!-- end list -->

````python
from scipy.stats import t

df_low = 3     # B·∫≠c t·ª± do th·∫•p (ƒëu√¥i n·∫∑ng)
df_medium = 10
df_high = 50   # B·∫≠c t·ª± do cao (g·∫ßn gi·ªëng normal)

x_t = np.linspace(-5, 5, 200)

plt.figure(figsize=(10, 6))
plt.plot(x_t, norm.pdf(x_t), 'k-', lw=2.5, alpha=0.8, label='Normal (Z)')
plt.plot(x_t, t.pdf(x_t, df=df_low), 'r--', lw=1.5, label=f't (df={df_low})')
plt.plot(x_t, t.pdf(x_t, df=df_medium), 'g-.', lw=1.5, label=f't (df={df_medium})')
plt.plot(x_t, t.pdf(x_t, df=df_high), 'b:', lw=1.5, label=f't (df={df_high})')
plt.legend()
plt.title('So s√°nh Ph√¢n ph·ªëi t v·ªõi Ph√¢n ph·ªëi Chu·∫©n t·∫Øc')
plt.xlabel('Gi√° tr·ªã')
plt.ylabel('M·∫≠t ƒë·ªô')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

## 10\. üî¨ ∆Ø·ªõc L∆∞·ª£ng Tham S·ªë (Fitting)

  * **M·ª•c ƒë√≠ch**: T√¨m c√°c tham s·ªë c·ªßa m·ªôt ph√¢n ph·ªëi l√Ω thuy·∫øt (v√≠ d·ª•: $\mu, \sigma$ c·ªßa `norm`) sao cho ph√¢n ph·ªëi ƒë√≥ kh·ªõp (fit) nh·∫•t v·ªõi d·ªØ li·ªáu m·∫´u ta c√≥.
  * **H√†m**: `<dist>.fit(data)` - Tr·∫£ v·ªÅ c√°c tham s·ªë ∆∞·ªõc l∆∞·ª£ng ƒë∆∞·ª£c.

<!-- end list -->

````python
# 1. T·∫°o d·ªØ li·ªáu m·∫´u (v√≠ d·ª•: t·ª´ N(15, 2^2))
np.random.seed(42) # ƒê·ªÉ k·∫øt qu·∫£ c√≥ th·ªÉ t√°i l·∫∑p
mu_true, sigma_true = 15, 2
data_sample = norm.rvs(loc=mu_true, scale=sigma_true, size=1000)

# 2. Gi·∫£ s·ª≠ kh√¥ng bi·∫øt mu, sigma. D√πng .fit() ƒë·ªÉ ∆∞·ªõc l∆∞·ª£ng t·ª´ data_sample
mu_fit, sigma_fit = norm.fit(data_sample)

print(f"Tham s·ªë g·ªëc:      mu = {mu_true:.4f}, sigma = {sigma_true:.4f}")
print(f"Tham s·ªë ∆∞·ªõc l∆∞·ª£ng: mu = {mu_fit:.4f}, sigma = {sigma_fit:.4f}")

# 3. V·∫Ω bi·ªÉu ƒë·ªì so s√°nh histogram c·ªßa d·ªØ li·ªáu v√† PDF v·ªõi tham s·ªë ∆∞·ªõc l∆∞·ª£ng
plt.figure(figsize=(10, 6))
plt.hist(data_sample, bins=30, density=True, alpha=0.6, label='Histogram D·ªØ li·ªáu M·∫´u')
x_fit = np.linspace(data_sample.min(), data_sample.max(), 200)
pdf_fit = norm.pdf(x_fit, loc=mu_fit, scale=sigma_fit)
plt.plot(x_fit, pdf_fit, 'r-', lw=2, label=f'PDF ∆Ø·ªõc l∆∞·ª£ng N({mu_fit:.2f}, {sigma_fit:.2f}^2)')
plt.title('∆Ø·ªõc l∆∞·ª£ng tham s·ªë Ph√¢n ph·ªëi Chu·∫©n t·ª´ D·ªØ li·ªáu M·∫´u')
plt.xlabel('Gi√° tr·ªã')
plt.ylabel('M·∫≠t ƒë·ªô')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

## üè≠ B√†i 2: Ph√¢n Ph·ªëi X√°c Su·∫•t (Scipy.stats)

B√†i n√†y t·∫≠p trung v√†o vi·ªác √°p d·ª•ng 3 ph√¢n ph·ªëi x√°c su·∫•t c∆° b·∫£n ƒë·ªÉ gi·∫£i quy·∫øt b√†i to√°n th·ª±c t·∫ø.

### (a) Ph√¢n Ph·ªëi Poisson (S·ªë l·∫ßn xu·∫•t hi·ªán trung b√¨nh)

  * [cite\_start]**M·ª•c ti√™u:** T√≠nh x√°c su·∫•t c√≥ **nhi·ªÅu h∆°n 3 l·ªói** ($P(X > 3)$), bi·∫øt trung b√¨nh l√† 2 l·ªói ($\lambda = 2$)[cite: 62, 63].
  * **Ph√¢n t√≠ch:** $P(X > 3) = 1 - P(X \le 3)$.
  * **H√†m:** S·ª≠ d·ª•ng `poisson.cdf` (H√†m ph√¢n ph·ªëi t√≠ch l≈©y).
      * `k = 3` (s·ªë l·ªói)
      * `mu = 2` (tham s·ªë $\lambda$ trong `scipy.stats` l√† `mu`)
  * **C√°ch l√†m:**

````python
# P(X > 3) = 1 - P(X <= 3)
prob_more_than_3 = 1 - poisson.cdf(3, mu=2)
print(f"X√°c su·∫•t c√≥ h∆°n 3 l·ªói: {prob_more_than_3:.4f}") # K·∫øt qu·∫£: 0.1429
````

### (b) Ph√¢n Ph·ªëi Nh·ªã Th·ª©c (S·ªë l·∫ßn th√†nh c√¥ng trong `n` ph√©p th·ª≠)

  * [cite\_start]**M·ª•c ti√™u:** T√≠nh x√°c su·∫•t c√≥ **ch√≠nh x√°c 18** vi m·∫°ch ƒë·∫°t chu·∫©n ($P(X = 18)$)[cite: 65].
  * **Ph√¢n t√≠ch:**
      * [cite\_start]S·ªë ph√©p th·ª≠ (t·ªïng s·ªë vi m·∫°ch): $n = 20$ [cite: 64]
      * [cite\_start]X√°c su·∫•t th√†nh c√¥ng (ƒë·∫°t chu·∫©n): $p = 0.95$ [cite: 64]
      * [cite\_start]S·ªë l·∫ßn th√†nh c√¥ng mong mu·ªën: $k = 18$ [cite: 65]
  * **H√†m:** S·ª≠ d·ª•ng `binom.pmf` (H√†m kh·ªëi x√°c su·∫•t - cho gi√° tr·ªã ch√≠nh x√°c).
  * **C√°ch l√†m:**

````python
prob_18_success = binom.pmf(18, n=20, p=0.95)
print(f"X√°c su·∫•t c√≥ ƒë√∫ng 18 vi m·∫°ch ƒë·∫°t: {prob_18_success:.4f}") # K·∫øt qu·∫£: 0.1887
````

### (c) Ph√¢n Ph·ªëi Chu·∫©n (Bi·∫øn li√™n t·ª•c)

  * [cite\_start]**M·ª•c ti√™u 1:** T√≠nh x√°c su·∫•t ho√†n th√†nh **trong v√≤ng 45 ph√∫t** ($P(X \le 45)$)[cite: 67].
  * **Ph√¢n t√≠ch:**
      * [cite\_start]Trung b√¨nh: $\mu = 50$ ph√∫t [cite: 66]
      * [cite\_start]ƒê·ªô l·ªách chu·∫©n: $\sigma = 5$ ph√∫t [cite: 66]
      * [cite\_start]Gi√° tr·ªã c·∫ßn t√≠nh: $x = 45$ [cite: 67]
  * **H√†m:** S·ª≠ d·ª•ng `norm.cdf` (H√†m ph√¢n ph·ªëi t√≠ch l≈©y - t√≠nh di·ªán t√≠ch b√™n tr√°i c·ªßa `x`).
  * **C√°ch l√†m (M·ª•c ti√™u 1):**

````python
# P(X <= 45) v·ªõi mu=50, sigma=5
prob_under_45 = norm.cdf(45, loc=50, scale=5)
print(f"X√°c su·∫•t ho√†n th√†nh trong 45 ph√∫t: {prob_under_45:.4f}") # K·∫øt qu·∫£: 0.1587
````

  * [cite\_start]**M·ª•c ti√™u 2:** T√¨m th·ªùi gian $x$ ƒë·ªÉ n·∫±m trong **top 5% nhanh nh·∫•t** ($P(X \le x) = 0.05$)[cite: 68].
  * **Ph√¢n t√≠ch:** ƒê√¢y l√† b√†i to√°n t√¨m ƒëi·ªÉm ph√¢n v·ªã (ng∆∞·ª£c c·ªßa CDF).
      * X√°c su·∫•t (quantile): $q = 0.05$ (v√¨ l√† 5% nhanh nh·∫•t, t·ª©c l√† 5% ·ªü ƒëu√¥i b√™n tr√°i)
  * **H√†m:** S·ª≠ d·ª•ng `norm.ppf` (H√†m ƒëi·ªÉm ph·∫ßn trƒÉm - Percent Point Function).
  * **C√°ch l√†m (M·ª•c ti√™u 2):**

````python
# T√¨m x sao cho P(X <= x) = 0.05
thresh_top_5 = norm.ppf(0.05, loc=50, scale=5)
print(f"Th·ªùi gian c·∫ßn ƒë·ªÉ v√†o top 5% nhanh nh·∫•t: {thresh_top_5:.2f} ph√∫t") # K·∫øt qu·∫£: 41.78 ph√∫t
````
