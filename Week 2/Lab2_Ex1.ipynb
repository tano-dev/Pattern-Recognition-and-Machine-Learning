**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
# **Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

# a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

# b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

# c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

#a


n = 10  # Số câu hỏi
p = 0.25  # Xác suất trả lời đúng mỗi câu hỏi
size = 10  # Số lần kiểm tra
correct_answers = binom.rvs(n, p, size=size)
print("Số câu đúng trong 10 lần kiểm tra:", correct_answers)

#b
# Tính số câu đúng cần thiết để được 4 điểm
score_needed = 4
# 4x - 2(10 - x) = 4 => 6x - 20 = 4 => 6x = 24 => x = 4
probability = binom.pmf(4, n, p)
print(f"Xác suất để sinh viên được {score_needed} điểm:", probability)

#c
expected_value = n * p
variance = n * p * (1 - p)
print("E(X):", expected_value)
print("Var(X):", variance)
# E(X) = x * p = 10 * 0.25 = 2.5
# Var(X) = n * p * (1 - p) = 10 * 0.25 * 0.75 = 1.875
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
# **Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

# 1 cuộc:
lambda_poisson = 3  # Số cuộc gọi trung bình mỗi phút
prob_1_call = poisson.pmf(1, lambda_poisson)
print("Xác suất nhận được 1 cuộc gọi trong một phút:", prob_1_call)
# 2 cuộc:
prob_2_calls = poisson.pmf(2, lambda_poisson)
print("Xác suất nhận được 2 cuộc gọi trong một phút:", prob_2_calls)
# 3 cuộc:
prob_3_calls = poisson.pmf(3, lambda_poisson)
print("Xác suất nhận được 3 cuộc gọi trong một phút:", prob_3_calls)
```

```python
# **Bài 03:** Trọng lượng (đơn vị gam) của một loại trái cây có phân phối chuẩn với µ = 500 (gam) gam và $σ^2 = 16 (gam^2)$. Trái cây thu hoạch được phân loại theo trọng lượng như sau:

# Loại 1: trên 505 gam

# Loại 2: từ 495 – 505 gam

# Loại 3: dười 495 gam

# a. Hãy tính tỷ lệ của mỗi loại.
# b. Bạn hãy mô phỏng lấy mẫu  20 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
# c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu
# d. Từ hai cách lấy mẫu ở câu b, c bạn có nhận xét gì không?


mu = 500  # Trung bình
sigma = 4  # Độ lệch chuẩn

#a
p_1 = 1 - norm.cdf(505, mu, sigma)
p_2 = norm.cdf(505, mu, sigma) - norm.cdf(495, mu, sigma)
p_3 = norm.cdf(495, mu, sigma)
print("Tỷ lệ loại 1 (trên 505 gam):", p_1)
print("Tỷ lệ loại 2 (từ 495 – 505 gam):", p_2)
print("Tỷ lệ loại 3 (dưới 495 gam):", p_3)
#b 
data_b = norm.rvs(loc=mu, scale=sigma, size=20)

dorongcot=1
offset = dorongcot/2  # Dùng để hiển thị ngay dấu ticks

plt.hist(data_b, bins=10, width=dorongcot)
plt.title("Histogram of Fruit Weights (Sample Size = 20)")
plt.xlabel("Weight (grams)")
plt.ylabel("Frequency")
plt.show()

mean_b = np.mean(data_b)
std_b = np.std(data_b)
print("Giá trị trung bình (mẫu 20):", mean_b)
print("Độ lệch chuẩn (mẫu 20):", std_b)

#c
data_c = norm.rvs(loc=mu, scale=sigma, size=100)
plt.hist(data_c, bins=10, width=dorongcot)
plt.title("Histogram of Fruit Weights (Sample Size = 100)")
plt.xlabel("Weight (grams)")
plt.ylabel("Frequency")
plt.show()
mean_c = np.mean(data_c)
std_c = np.std(data_c)
print("Giá trị trung bình (mẫu 100):", mean_c)
print("Độ lệch chuẩn (mẫu 100):", std_c)

#d
# Nhận xét: Khi kích thước mẫu tăng lên, giá trị trung bình và độ lệch chuẩn của mẫu càng gần với giá trị thực của phân phối chuẩn ban đầu (µ = 500, σ = 4). Điều này minh họa định lý giới hạn trung tâm, cho thấy rằng với kích thước mẫu lớn, phân phối mẫu sẽ xấp xỉ phân phối chuẩn.
```

```python
# **Bài 05:** Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

# a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

# b. Vẽ đồ thị boxplot

# c. Vẽ các đồ thị: tần số, tần suất

data = binom.rvs(n=50, p=0.7, size=5000)
# a
min_value = np.min(data)
max_value = np.max(data)
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)
Q3 = np.percentile(data, 75)

print("Min:", min_value)
print("Max:", max_value)
print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
# b
plt.boxplot(data)
plt.title("Boxplot of Binomial Distribution (n=50, p=0.7)")
plt.ylabel("Value")
plt.show()
# c
# Tần số
plt.hist(data, bins=30, edgecolor='black')
plt.title("Frequency Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Tần suất
plt.hist(data, bins=30, density=True, edgecolor='black')
plt.title("Relative Frequency Histogram")
plt.xlabel("Value")
plt.ylabel("Relative Frequency")
plt.show()
```

```python
# **Bài 6**

# Giả sử X là biến ngẫu nhiên rời rạc

# a. Hãy cho biết ý nghĩa câu lệnh sau: `binom.pmf(4, 10, 0.6)`

# b. Giả sử X có phân phối nhị thức `X ~ B(10, 0.6)`. Hãy tính các giá trị sau:  
# $P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

# c. Hãy cho biết ý nghĩa câu lệnh sau: `poisson.pmf(4, 3)`

# d. Giả sử X có phân phối Poisson `X ~ P(3)`. Hãy tính các giá trị sau:  
# $P(X \leq 5), P(X \le 5), P(X \ge 4), P(X = 5) $

#a
# Ý nghĩa: Tính xác suất để biến ngẫu nhiên X có phân phối nhị thức với n=10 và p=0.6 nhận giá trị đúng bằng 4.

#b
n = 10
p = 0.6
data = binom(n, p)
P_X_le_5 = data.cdf(5)
P_X_ge_4 = 1 - data.cdf(3)
P_X_eq_5 = data.pmf(5)
print("P(X ≤ 5):", P_X_le_5)
print("P(X ≥ 4):", P_X_ge_4)
print("P(X = 5):", P_X_eq_5)
#c
# Ý nghĩa: Tính xác suất để biến ngẫu nhiên X có phân phối Poisson với λ=3 nhận giá trị đúng bằng 4.
```

**Bài 7**

a. Cho biết ý nghĩa của các câu lệnh sau:

- `norm.cdf(2)`

- `norm.cdf(2, 1, 1)`

b. Cho biết kết quả của câu lệnh sau: `norm.ppf(norm.cdf(2))`

c. Tính các giá trị sau và vẽ hình minh họa cho các giá trị tính được:

- `norm.ppf(0.975)`

- `norm.ppf(0.975, 1, 1)`

- `norm.ppf(0.975, 1, 2)`

d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

- $P(X \leq 20)$
- $P(X \ge 29.5)$
- $P(X = 23.8)$

```python
# **Bài 7**

# a. Cho biết ý nghĩa của các câu lệnh sau:

# - `norm.cdf(2)`

# - `norm.cdf(2, 1, 1)`

# - `norm.cdf(2, 1, 2)`

# b. Cho biết kết quả của câu lệnh sau: `norm.ppf(norm.cdf(2))`

# c. Tính các giá trị sau và vẽ hình minh họa cho các giá trị tính được:

# - `norm.ppf(0.975)`

# - `norm.ppf(0.975, 1, 1)`

# - `norm.ppf(0.975, 1, 2)`

# d. Giả sử biến ngẫu nhiên X có phân phối chuẩn với kỳ vọng là 24 và phương sai là 16. Hãy tính các giá trị sau:

# - $P(X \leq 20)$
# - $P(X \ge 29.5)$
# - $P(X = 23.8)$

#a
# - `norm.cdf(2)`: Tính hàm phân phối tích lũy (CDF) của phân phối chuẩn chuẩn hóa tại điểm 2.
# - `norm.cdf(2, 1, 1)`: Tính CDF của phân phối chuẩn với trung bình 1 và độ lệch chuẩn 1 tại điểm 2.
# - `norm.cdf(2, 1, 2)`: Tính CDF của phân phối chuẩn với trung bình 1 và độ lệch chuẩn 2 tại điểm 2.

#b
result_b = norm.ppf(norm.cdf(2))
print("Kết quả của norm.ppf(norm.cdf(2)):", result_b)

#c 
ppf_975 = norm.ppf(0.975)
ppf_975_1_1 = norm.ppf(0.975, 1, 1)
ppf_975_1_2 = norm.ppf(0.975, 1, 2)
print("norm.ppf(0.975):", ppf_975)
print("norm.ppf(0.975, 1, 1):", ppf_975_1_1)
print("norm.ppf(0.975, 1, 2):", ppf_975_1_2)
# Vẽ hình minh họa
x = np.linspace(-4, 4, 1000)
plt.plot(x, norm.pdf(x), label='Standard Normal PDF')
plt.axvline(ppf_975, color='r', linestyle='--', label='norm.ppf(0.975)')
plt.axvline(ppf_975_1_1, color='g', linestyle='--', label='norm.ppf(0.975, 1, 1)')
plt.axvline(ppf_975_1_2, color='b', linestyle='--', label='norm.ppf(0.975, 1, 2)')
plt.title("Illustration of norm.ppf values")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()
#d
mu = 24
sigma = 4  # Độ lệch chuẩn là căn bậc hai của phương
# - P(X ≤ 20)
P_X_le_20 = norm.cdf(20, mu, sigma)
print("P(X ≤ 20):", P_X_le_20)
# - P(X ≥ 29.5)
P_X_ge_29_5 = 1 - norm.cdf(29.5, mu, sigma)
print("P(X ≥ 29.5):", P_X_ge_29_5)
# - P(X = 23.8)
P_X_eq_23_8 = norm.pdf(23.8, mu, sigma)
print("P(X = 23.8):", P_X_eq_23_8)
```
