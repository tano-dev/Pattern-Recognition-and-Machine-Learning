## üî¢ 2. M·∫£ng NumPy (ndarray)
* **ndarray**: L√† t·∫≠p h·ª£p c√°c ph·∫ßn t·ª≠ **c√πng ki·ªÉu d·ªØ li·ªáu**.
* **rank**: S·ªë chi·ªÅu c·ªßa m·∫£ng.

* Truy c·∫≠p ph·∫ßn t·ª≠: D√πng d·∫•u ngo·∫∑c vu√¥ng (vd: `a[0]`, `b[0, 1]`).

## ‚ú® 3. H√†m T·∫°o M·∫£ng

* `np.full((rows, cols), value)`: T·∫°o m·∫£ng v·ªõi t·∫•t c·∫£ ph·∫ßn t·ª≠ l√† `value`.
* `np.eye(n)`: T·∫°o ma tr·∫≠n ƒë∆°n v·ªã k√≠ch th∆∞·ªõc `n x n`.
* `np.arange(start, stop, step)`: T·∫°o m·∫£ng c√°c s·ªë c√°ch ƒë·ªÅu nhau (t∆∞∆°ng t·ª± `range`).
* `np.linspace(start, stop, num)`: T·∫°o m·∫£ng g·ªìm `num` s·ªë c√°ch ƒë·ªÅu nhau t·ª´ `start` ƒë·∫øn `stop`.
* `np.random.random((rows, cols))`: T·∫°o m·∫£ng v·ªõi c√°c gi√° tr·ªã ng·∫´u nhi√™n (t·ª´ 0 ƒë·∫øn 1).

## üéØ 4. Ch·ªâ M·ª•c (Indexing)
* **Slicing**: T∆∞∆°ng t·ª± list, nh∆∞ng c√≥ th·ªÉ √°p d·ª•ng cho nhi·ªÅu chi·ªÅu.
    * `a[:2, 1:3]`: L·∫•y 2 d√≤ng ƒë·∫ßu, c·ªôt th·ª© 1 v√† 2 (kh√¥ng bao g·ªìm c·ªôt 3).
    * Slice l√† m·ªôt **view** (tham chi·∫øu) t·ªõi d·ªØ li·ªáu g·ªëc, thay ƒë·ªïi slice s·∫Ω thay ƒë·ªïi m·∫£ng g·ªëc.
    * K·∫øt h·ª£p s·ªë nguy√™n v√† slice: `a[1, :]` (l·∫•y d√≤ng th·ª© 1, rank gi·∫£m 1), `a[1:2, :]` (l·∫•y d√≤ng th·ª© 1, rank gi·ªØ nguy√™n).
* **Integer Indexing**: Ch·ªçn c√°c ph·∫ßn t·ª≠ kh√¥ng li√™n ti·∫øp.
    * `a[[0, 1, 2], [0, 1, 0]]`: L·∫•y c√°c ph·∫ßn t·ª≠ `a[0,0]`, `a[1,1]`, `a[2,0]`.
    * C√≥ th·ªÉ l·∫∑p l·∫°i ch·ªâ s·ªë: `a[[0, 0], [1, 1]]` (l·∫•y `a[0,1]` hai l·∫ßn).
    * Ch·ªçn/thay ƒë·ªïi 1 ph·∫ßn t·ª≠ m·ªói h√†ng: `a[np.arange(4), [0, 2, 0, 1]]`.
* **Boolean Indexing**: Ch·ªçn ph·∫ßn t·ª≠ th·ªèa m√£n ƒëi·ªÅu ki·ªán.
    * `bool_idx = (a > 2)`: T·∫°o m·∫£ng boolean c√πng shape v·ªõi `a`.
    * `a[bool_idx]`: L·∫•y c√°c ph·∫ßn t·ª≠ ·ª©ng v·ªõi gi√° tr·ªã `True`.
    * Vi·∫øt t·∫Øt: `a[a > 2]`.

## üßÆ 5. Ph√©p To√°n Tr√™n M·∫£ng

# Ph√¢n Ph·ªëi X√°c Su·∫•t

## 1. Gi·ªõi Thi·ªáu
* **Ph√¢n Ph·ªëi X√°c Su·∫•t (PPXS)**: Th·ªÉ hi·ªán x√°c su·∫•t xu·∫•t hi·ªán c·ªßa c√°c gi√° tr·ªã c·ªßa m·ªôt bi·∫øn ng·∫´u nhi√™n.
* **M·ª•c ƒë√≠ch**: Hi·ªÉu quy lu·∫≠t c·ªßa d·ªØ li·ªáu, d·ª± ƒëo√°n gi√° tr·ªã, so s√°nh d·ªØ li·ªáu v·ªõi c√°c PPXS ƒë√£ bi·∫øt.

---

## 2. Th∆∞ Vi·ªán `scipy.stats`
* Th∆∞ vi·ªán ch√≠nh trong Python ƒë·ªÉ l√†m vi·ªác v·ªõi c√°c PPXS.
* M·ªôt s·ªë ph√¢n ph·ªëi th√¥ng d·ª•ng v√† c√°ch import:

| Scipy Package | Distribution Name         | T√™n Ph√¢n Ph·ªëi             | K√Ω hi·ªáu                     | C√°ch Import                 |
| ------------- | ------------------------- | ------------------------- | --------------------------- | --------------------------- |
| `binom`       | Binomial Distribution     | Ph√¢n Ph·ªëi Nh·ªã Th·ª©c        | $B(n, p)$                   | `from scipy.stats import binom` |
| `poisson`     | Poisson Distribution      | Ph√¢n Ph·ªëi Poisson         | $P(\lambda)$                | `from scipy.stats import poisson`|
| `norm`        | Normal/Gauss Distribution | Ph√¢n Ph·ªëi Chu·∫©n           | $N(\mu, \sigma^2)$          | `from scipy.stats import norm`  |
| `t`           | Student Distribution      | Ph√¢n Ph·ªëi Student         | $T$                         | `from scipy.stats import t`     |
| `chi2`        | Chi Square Distribution   | Ph√¢n Ph·ªëi Chi B√¨nh Ph∆∞∆°ng | $\chi^2$                    |                             |

## 3. C√°c H√†m Th√¥ng D·ª•ng trong `scipy.stats`
* `<distribution>.rvs(...)`: T·∫°o m·∫´u ng·∫´u nhi√™n (Random Variates).
    * `size`: S·ªë l∆∞·ª£ng m·∫´u c·∫ßn t·∫°o.
    * C√°c tham s·ªë ƒë·∫∑c tr∆∞ng c·ªßa ph√¢n ph·ªëi (vd: `n`, `p` cho `binom`; `loc`, `scale` cho `norm`).
* `<distribution>.pmf(k, ...)`: H√†m kh·ªëi x√°c su·∫•t (Probability Mass Function) - $P(X=k)$ cho ph√¢n ph·ªëi **r·ªùi r·∫°c**.
* `<distribution>.pdf(x, ...)`: H√†m m·∫≠t ƒë·ªô x√°c su·∫•t (Probability Density Function) - $f(x)$ cho ph√¢n ph·ªëi **li√™n t·ª•c**.
* `<distribution>.cdf(x, ...)`: H√†m ph√¢n ph·ªëi t√≠ch l≈©y (Cumulative Distribution Function) - $P(X \le x)$.
* `<distribution>.ppf(q, ...)`: H√†m ƒëi·ªÉm ph·∫ßn trƒÉm (Percent Point Function) - H√†m ng∆∞·ª£c c·ªßa CDF, t√¨m $x$ sao cho $P(X \le x) = q$.
* `<distribution>.sf(x, ...)`: H√†m s·ªëng s√≥t (Survival Function) - $P(X > x)$ (b·∫±ng `1 - cdf(x)`).
* `<distribution>.isf(q, ...)`: H√†m ng∆∞·ª£c c·ªßa h√†m s·ªëng s√≥t - t√¨m $x$ sao cho $P(X > x) = q$.

---

## 4. Ph√¢n Ph·ªëi Nh·ªã Th·ª©c (`binom`)
* **K√Ω hi·ªáu**: $B(n, p)$
    * `n`: S·ªë l·∫ßn th·ª±c hi·ªán ph√©p th·ª≠ Bernoulli ƒë·ªôc l·∫≠p.
    * `p`: X√°c su·∫•t th√†nh c√¥ng trong m·ªôt l·∫ßn th·ª≠.
* **Bi·∫øn ng·∫´u nhi√™n X**: S·ªë l·∫ßn th√†nh c√¥ng trong `n` ph√©p th·ª≠.
* **V√≠ d·ª•**: S·ªë c√¢u ƒë√∫ng khi ƒë√°nh l·ª•i 10 c√¢u tr·∫Øc nghi·ªám (4 ch·ªçn 1). `n=10`, `p=0.25`.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

n, p = 10, 0.25

# T·∫°o 1 m·∫´u ng·∫´u nhi√™n
mau_1 = binom.rvs(n, p)

# T·∫°o 1000 m·∫´u ng·∫´u nhi√™n
mau_1000 = binom.rvs(n, p, size=1000)

# V·∫Ω histogram
plt.hist(mau_1000, bins=np.arange(n+2)-0.5, width=0.4) # Ch·ªânh bins v√† width
plt.show()

# T√≠nh v√† v·∫Ω h√†m PMF (x√°c su·∫•t t·∫°i t·ª´ng gi√° tr·ªã k=0..10)
k_values = np.arange(n + 1)
pmf_values = binom.pmf(k_values, n, p)
plt.scatter(k_values, pmf_values, color='red')
plt.plot(k_values, pmf_values, color='orange', ls='--') # N·ªëi ƒëi·ªÉm cho d·ªÖ nh√¨n
plt.show()

# T√≠nh v√† v·∫Ω h√†m CDF (x√°c su·∫•t t√≠ch l≈©y P(X <= k))
cdf_values = binom.cdf(k_values, n, p)
plt.plot(k_values, cdf_values, marker='o')
plt.show() 
```

## 4. Ph√¢n Ph·ªëi Chu·∫©n (`norm`)
* **K√Ω hi·ªáu**: $N(\mu, \sigma^2)$
    * ($\mu$): K·ª≥ v·ªçng (trung b√¨nh).
    *  ($\sigma$): ƒê·ªô l·ªách chu·∫©n.
* **Bi·∫øn ng·∫´u nhi√™n X**:  Bi·∫øn li√™n t·ª•c, ƒë·ªì th·ªã h√¨nh chu√¥ng ƒë·ªëi x·ª©ng qua $\mu$.
* **V√≠ d·ª•**: Chi·ªÅu cao thanh ni√™n ($\mu=170$ cm, $\sigma=3$ cm).


```python 
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 170, 3

# T·∫°o 10000 m·∫´u ng·∫´u nhi√™n
```
