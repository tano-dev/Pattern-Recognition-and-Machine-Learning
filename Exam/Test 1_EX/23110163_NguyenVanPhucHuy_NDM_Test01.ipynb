```python
# THỐNG KÊ MÁY TÍNH & ỨNG DỤNG

# Bài 1: Cho dataset mpg.csv về mức tiêu hao nhiên liệu. Thông tin của dữ liệu như sau:

# mpg: Miles/(US) gallon số gallon xăng trên 1 dặm
# cylinders: số lượng xilanh
# displacement: Displacement/ dung tích xi lanh (cu.in / inches3)
# horsepower: công suất theo mã lực
# weight: trọng lượng (pound)
# acceleration: Gia tốc
# model_year: Năm sản xuất
# origin: Xuất xứ
# name: Tên dòng xe
# Hãy thực hiện:

# (1.0đ) Đọc file dữ liệu và kiểm tra xem có dòng nào bị thiếu dữ liệu không? Nếu có, hãy xóa những dòng đó. Có dữ liệu nào bị trùng lặp không? Nếu có, xử lý thế nào?
# (1.0đ) Tính giá trị trung bình và trung vị của các thuộc tính: 'cylinders', 'horsepower', 'weight'. Tìm miền giá trị (min, max) và phân vị bậc (Q1, Q3) của 'mpg'. Có xe nào có mức tiêu hao nhiên liệu đặc biệt cao hoặc thấp không?
# (1.0đ) Tính mức tiêu hao nhiên liệu trung bình (mpg) của từng nhóm xe dựa trên 'origin'. Xe từ khu vực nào có xu hướng tiêu hao ít nhiên liệu nhất?
# (1.0đ) Vẽ Histogram thể hiện sự phân bố của trọng lượng (weight) các xe. So sánh Boxplot của công suất (horsepower) giữa xe sản xuất trong thập niên 70 và 80. Xe từ thập niên nào có công suất cao hơn? Nhận xét xu hướng.
# (1.0đ) Vẽ Scatter Plot thể hiện mối quan hệ giữa 'mpg' với các yếu tố: Số lượng xi-lanh (cylinders), Công suất (horsepower), Gia tốc (acceleration), Năm sản xuất (model_year). Yếu tố nào có ảnh hưởng lớn nhất đến mức tiêu hao nhiên liệu?


# Bài 2: Một nhà máy sản xuất vi mạch điện tử đang nghiên cứu về tỷ lệ lỗi trong sản phẩm, số lượng sản phẩm đạt tiêu chuẩn, cũng như thời gian hoàn thành mỗi lô hàng. Quá trình sản xuất được mô hình hóa bằng các phân phối xác suất phù hợp.
# (a) (1,5 điểm) Trung bình mỗi lô hàng có 2 lỗi sản xuất. Giả sử số lỗi tuân theo phân phối Poisson. Tính xác suất một lô hàng nhiều hơn 3 lỗi.
# (b) (1,5 điểm) Mỗi lô hàng gồm 20 vi mạch, và xác suất mỗi vi mạch đạt tiêu chuẩn là 95%. Tính xác suất một lô hàng có chính xác 18 vi mạch đạt tiêu chuẩn.
# (c) (2 điểm) Thời gian hoàn thành một lô hàng tuân theo phân phối chuẩn, với trung bình 50 phút và độ lệch chuẩn 5 phút.
# Tính xác suất một lô hàng được hoàn thành trong vòng 45 phút.
# Để nằm trong top 5% lô hàng sản xuất nhanh nhất, cần hoàn thành trong thời gian bao nhiêu?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson, binom, norm

# Bài 1
# Đọc file dữ liệu và kiểm tra dữ liệu thiếu và trùng lặp
df = pd.read_csv('mpg.csv')
print("Kiểm tra dữ liệu thiếu:")
print(df.isnull().sum())
df = df.dropna()
print("\nKiểm tra dữ liệu trùng lặp:")
print(df.duplicated().sum())
df = df.drop_duplicates()
print("\nDữ liệu sau khi xử lý thiếu và trùng lặp:")
print(df.info())

# Tính giá trị trung bình và trung vị của các thuộc tính: 'cylinders', 'horsepower', 'weight'.
print("\nTính giá trị trung bình và trung vị:")
print("Giá trị trung bình:")
print(df[['cylinders', 'horsepower', 'weight']].mean())
print("Giá trị trung vị:")
print(df[['cylinders', 'horsepower', 'weight']].median())

# Tìm miền giá trị (min, max) và phân vị bậc (Q1, Q3) của 'mpg'. Có xe nào có mức tiêu hao nhiên liệu đặc biệt cao hoặc thấp không?
mpg_min = df['mpg'].min()
mpg_max = df['mpg'].max()
mpg_q1 = df['mpg'].quantile(0.25)
mpg_q3 = df['mpg'].quantile(0.75)
mpg_range = (mpg_min, mpg_max)
print(f"\nMiền giá trị của MPG: {mpg_range}")
print(f"Phân vị Q1, Q3 của MPG: ({mpg_q1}, {mpg_q3})")
print("\nXe có mức tiêu hao nhiên liệu đặc biệt cao hoặc thấp:")
outliers = df[(df['mpg'] < mpg_q1 - 1.5 * (mpg_q3 - mpg_q1)) | (df['mpg'] > mpg_q3 + 1.5 * (mpg_q3 - mpg_q1))]
print("Các xe có mức tiêu hao nhiên liệu đặc biệt:")
print(outliers[['name', 'mpg']])

# Tính mức tiêu hao nhiên liệu trung bình (mpg) của từng nhóm xe dựa trên 'origin'. Xe từ khu vực nào có xu hướng tiêu hao ít nhiên liệu nhất?
mpg_by_origin = df.groupby('origin')['mpg'].mean()
print("\nMức tiêu hao nhiên liệu trung bình (mpg) của từng nhóm xe dựa trên 'origin':")
print(mpg_by_origin)
least_consuming_origin = mpg_by_origin.idxmin()
print(f"Xe từ khu vực {least_consuming_origin} có xu hướng tiêu hao ít nhiên liệu nhất.")

# Vẽ Histogram thể hiện sự phân bố của trọng lượng (weight) các xe.
plt.figure(figsize=(10, 6))
sns.histplot(df['weight'], bins=30, kde=True)
plt.title('Phân bố trọng lượng (weight) các xe')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
# So sánh Boxplot của công suất (horsepower) giữa xe sản xuất trong thập niên 70 và 80.
df['decade'] = df['model_year'].apply(lambda x: '70s' if 70 <= x < 80 else ('80s' if 80 <= x < 90 else 'Other'))
plt.figure(figsize=(10, 6))
sns.boxplot(x='decade', y='horsepower', data=df[df['decade'] != 'Other'])
plt.title('So sánh Boxplot của công suất (horsepower) giữa xe sản xuất trong thập niên 70 và 80')
plt.xlabel('Decade')
plt.ylabel('Horsepower')
plt.show()
print("\nXe từ thập niên nào có công suất cao hơn?")
hp_70s = df[df['decade'] == '70s']['horsepower'].median()
hp_80s = df[df['decade'] == '80s']['horsepower'].median()
if hp_70s > hp_80s:
    print("Xe từ thập niên 70 có công suất cao hơn.")
else:
    print("Xe từ thập niên 80 có công suất cao hơn.")
print("Nhận xét xu hướng: Công suất xe có xu hướng tăng theo thời gian do cải tiến công nghệ và yêu cầu thị trường.")

# Vẽ Scatter Plot thể hiện mối quan hệ giữa 'mpg' với các yếu tố: Số lượng xi-lanh (cylinders), Công suất (horsepower), Gia tốc (acceleration), Năm sản xuất (model_year).
factors = ['cylinders', 'horsepower', 'acceleration', 'model_year']
for factor in factors:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=factor, y='mpg', data=df)
    plt.title(f'Mối quan hệ giữa mpg và {factor}')
    plt.xlabel(factor)
    plt.ylabel('mpg')
    plt.show()
    correlation = df['mpg'].corr(df[factor])
    print(f"Hệ số tương quan giữa mpg và {factor}: {correlation}")
# Yếu tố nào có ảnh hưởng lớn nhất đến mức tiêu hao nhiên liệu?
correlations = {factor: df['mpg'].corr(df[factor]) for factor in factors}
most_influential_factor = max(correlations, key=lambda k: abs(correlations[k]))
print(f"\nYếu tố có ảnh hưởng lớn nhất đến mức tiêu hao nhiên liệu là: {most_influential_factor} với hệ số tương quan {correlations[most_influential_factor]}")

# Bài 2
# (a) Trung bình mỗi lô hàng có 2 lỗi sản xuất. Giả sử số lỗi tuân theo phân phối Poisson. Tính xác suất một lô hàng nhiều hơn 3 lỗi.
lambda_poisson = 2
prob_more_than_3_errors = 1 - poisson.cdf(3, lambda_poisson)
print(f"\nXác suất một lô hàng nhiều hơn 3 lỗi: {prob_more_than_3_errors}")

# (b) Mỗi lô hàng gồm 20 vi mạch, và xác suất mỗi vi mạch đạt tiêu chuẩn là 95%. Tính xác suất một lô hàng có chính xác 18 vi mạch đạt tiêu chuẩn.
n = 20
p = 0.95
k = 18
prob_exactly_18 = binom.pmf(k, n, p)
print(f"\nXác suất một lô hàng có chính xác 18 vi mạch đạt tiêu chuẩn: {prob_exactly_18}")

# (c) Thời gian hoàn thành một lô hàng tuân theo phân phối chuẩn, với trung bình 50 phút và độ lệch chuẩn 5 phút.
mu = 50
sigma = 5
# Tính xác suất một lô hàng được hoàn thành trong vòng 45 phút.
prob_within_45 = norm.cdf(45, mu, sigma)
print(f"\nXác suất một lô hàng được hoàn thành trong vòng 45 phút: {prob_within_45}")

# Để nằm trong top 5% lô hàng sản xuất nhanh nhất, cần hoàn thành trong thời gian bao nhiêu?
time_to_top_5 = norm.ppf(0.95, mu, sigma)
print(f"\nĐể nằm trong top 5% lô hàng sản xuất nhanh nhất, cần hoàn thành trong thời gian: {time_to_top_5}")
```
